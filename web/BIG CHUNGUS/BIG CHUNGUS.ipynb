## Analysis
Upon browsing the site, we're presented with this very loud site (with the image in the middle spinning)



On inspection of index.js, I found a line of code
```js
if (req.query.username.length > 0xB16_C4A6A5) {
```

First thoughts are that one appears to have to send a username approximately 47GB long - not likely.

In the absence of any other obvious defects, I thought about whether one could give a query string that would cause that to be the case. The obvious first thing to check would be ?username.length=0xb16c4a6a6 which didn't succeed. I remembered back to my web dev days and recalled that one could (at least in PHP) specify an array selector to set the values ?username[length]=0xb16c4a6a6

On submitting this query string, chungus detected
