Let's try to make this a bit cleaner

```c
char* DAT_0x0105560 = "What's your favorite number? :D";
puts(DAT_0x0105560)
char[16] DAT_0x0105540;
read(0, DAT_0x0105540, 16);
char* newline = strchr(DAT_0x0105540, '\n')
if (newline != 0) {
    newline = 0;
}

ulonglong DAT_00105560  = 0x6E8EEA35727D5CE8A297FA50CC11;
ulonglong DAT_0x105540 ^= 0x20EB9C06215D25BCCCA48C35FD54;
strcmp(DAT_0x105540, DAT_0x105560)
```

Hence the favorite number is 0x20EB9C06215D25BCCCA48C35FD54 ^ 0x6E8EEA35727D5CE8A297FA50CC11

````python
favoriteNumber = bytes(reversed((0x20EB9C06215D25BCCCA48C35FD54 ^ 0x6E8EEA35727D5CE8A297FA50CC11).to_bytes(16)))[:-2]

from pwn import *
conn = process("./lonely_bot")
conn.recvlineS()
conn.sendline(favoriteNumber)
print(conn.recvallS())
````
