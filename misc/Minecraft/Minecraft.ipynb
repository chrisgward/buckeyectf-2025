Now for the fun task of reading off the binary. Lever up = 1, down = 0

``` 10011101 10011100 10001011 10011001 10000100 11000111 11001111 11001111 11001110 11001100 11001011 10110001 11001110 11001111 11001001 11001110 10111100 10000010```

The first bit always appears to be 1. I suspect this is ASCII, and I picked the wrong bits. Let's ingest it in Python and then manipulate further.


````python
bits = "10011101 10011100 10001011 10011001 10000100 11000111 11001111 11001111 11001110 11001100 11001011 10110001 11001110 11001111 11001001 11001110 10111100 10000010"
bs = bits.split()

for byte in bs:
    print(chr(0xFF - int(byte, 2)), end="")
````
