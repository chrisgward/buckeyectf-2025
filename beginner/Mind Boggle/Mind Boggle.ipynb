My friend gave me this file, but I have no idea what it means!!!!! HELP

## Included files
* mystery.txt

## Analysis
On opening the file, I recognise the contents to be [Brainfuck](https://en.wikipedia.org/wiki/Brainfuck) - an esoteric programming language.

## Solution
Let's execute it using [Python-Brainfuck by pocmo](https://github.com/pocmo/Python-Brainfuck/blob/master/brainfuck.py)
I modified this to return a string instead of printing to stdout.

```python
import brainfuck
output = brainfuck.execute("mystery.txt")
print(output)
```

As the result is encoded in hex, let's decode this

```python
from pwn import *
unhex(output)
```

And again, another encoding, looks vaguely base64

```python
b64d(unhex(output)).decode()
```
