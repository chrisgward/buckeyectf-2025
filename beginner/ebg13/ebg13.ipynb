## Analysis
Browsing the website, we are presented with the following menu

Original:

Upon inspection of the source code, I see that the inline stylesheet is also being rotated.

Because this application is acting as a proxy, it is vulnerable to a confused-deputy attack.

Upon inspection of the source code, we see that there is a /admin endpoint, restricted to localhost, which prints the flag, and we are presented with an error externally.

![denied.png](denied.png)

If we ask for the url of http://localhost:3000/admin, it will be proxied locally, but will be ROT13'd.

![flag_rot13.png](flag_rot13.png)
